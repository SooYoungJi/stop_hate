### Clean 약 1만개 set 만들기

```python
#혐오감정의 문장들은 혐오표현이 들어있을 가능성이 크므로 일단 제외
tmp_drop_one = tmp[tmp['Emotion']!='혐오']

#행복은 클린 문장으로 분류가 쉬운 데이터일거라 가정하고 일단 제외
tmp_drop_two = tmp_drop_one[tmp_drop_one['Emotion']!='행복']
tmp_drop_two.Emotion.value_counts()
```

중립, 슬픔, 놀람, 분노, 공포 5개 감정에서 각 랜덤으로 2100개씩 뽑아서 10500개를 excel로 확인해서 혐오표현은 제거함.

```python
tmp_drop_two['Emotion'].unique()
```

```python
sample_df = pd.DataFrame(columns=['Sentence','Emotion']) ## 데이터 초기값

for g in tmp_drop_two['Emotion'].unique():
    temp_df = tmp_drop_two.query('Emotion==@g').sample(n=2100) ## 그룹별 데이터 추출 및 2개 비복원 추출
    sample_df = pd.concat([sample_df,temp_df]) ## 데이터 추가

sample_df = sample_df.reset_index(drop=True)
```
